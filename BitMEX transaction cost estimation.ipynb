```python

ph = bitcoind.getblockchaininfo()

def brekt(height):
    blk = bitcoind.getblock(bitcoind.getblockhash(height), 2)
    n,b,tn,tb=(0,0,0,0)

    coinbase_value = 0
    for vout in blk['tx'][0]['vout']:
        coinbase_value += vout['value']
        
    for tx in blk['tx']:
        if bitmex_keys[1] in tx['hex']:
            n += 1
            b += len(tx['hex'])/2
        tn += 1
        tb += len(tx['hex'])/2
    return [n,b,tn,tb, coinbase_value]


data = {}
for h in range(543231, 543853):
    if not h in data:
        data[h] = brekt(h)
    
```
