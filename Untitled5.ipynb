```python
def fgentle(x, ceil=0):
    if x == 0: return 1
    p = 2
    ex = np.exp(x/100)
    if ceil == 0:
        if ex > 10: p=1
        if ex > 100: p=0
    else:
        ex = np.ceil(ex)
        p=0
        
    return ('{:.'+str(p)+'f}').format(ex)
fgentle0 = functools.partial(fgentle, ceil=0)
fgentle1 = functools.partial(fgentle, ceil=1)
```
