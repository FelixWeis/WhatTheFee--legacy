```python
import pathlib
import multiprocessing
import subprocess

l = list(sorted(pathlib.Path('dataset-blocks').glob('blk*.tsv')))
```

```python
import multiprocessing
```

```python
def func(f):
    
    print(f'{f.name}: starting')
    process = subprocess.Popen('python worker.py ' + f.name, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    print(process.stderr.read())
    print(f'{f.name}: done')

with multiprocessing.Pool() as p:
    p.map(func, l)
    
```
