```python

def prepare_folder(folder, ceilgroup=40):
    PATH_PREFIX=folder
    
    txpool_mpool = pd.concat(wtfutil.pqload_lastn(PATH_PREFIX, 'full', 18).values())
    feeagg_mpool = wtfutil.feeagg(txpool_mpool, FEEAGG_BUCKETS) / 18
    
    txpool_delta = pd.concat(wtfutil.pqload_lastn(PATH_PREFIX, 'diff', DELTA_DATAPOINTS)).query('diff == 1')
    feeagg_delta = wtfutil.feeagg(txpool_delta, FEEAGG_BUCKETS) * 40 / DELTA_DATAPOINTS
    
    fa_demo = feeagg_mpool
    
    fn = functools.partial(wtfutil.cellfn_simulate, fa_start=fa_demo, fa_delta=feeagg_delta, mvs=1.0)
    r=wtfutil.mp_map_df(fn, wtfutil.df_simulation_large2)
    
    print()
    return (
        '-'.join(folder.split('/')[1:]),
        wtfutil.fa_ceilgroup(feeagg_mpool).mvs, 
        wtfutil.fa_ceilgroup(feeagg_delta).mvs, 
        r)

    
```

```python
all_mpools = dict()
all_deltas = dict()
all_feetbl = dict()
```

```python
for folder in folders:
    print(folder)
    s = prepare_folder(folder)
    print(s[0])
    all_mpools[s[0]] = s[1]
    all_deltas[s[0]] = s[2]
    all_feetbl[s[0]] = s[3]
```

```python
prios = pd.Series([[3,1], [9,6], [8, 18], [72,86], [1,0]])

def solve_prio(x, vals_mpool, vals_delta):
    s = wtfutil.simulate_simple(vals_mpool, vals_delta, nmeans=x[0], nblocks=x[1])
    if not len(s):
        return -1
    isol = s[-1][-1]
    return vals_mpool.index[isol]

simple = pd.Series([solve_prio(x, feeagg_mpool.mvs, feeagg_delta.mvs) for x in prios])

pd.DataFrame(simple, columns=['feelvl'])
```

```python
lvls = prios.map(functools.partial(solve_prio, 
                            vals_mpool=feeagg_mpool.mvs, 
                            vals_delta=feeagg_delta.mvs))

lvls_slice = [
    feeagg_delta.index >= lvls[0],
    (feeagg_delta.index >= lvls[1]) & (feeagg_delta.index < lvls[0]),
    (feeagg_delta.index >= lvls[2]) & (feeagg_delta.index < lvls[1]),
    (feeagg_delta.index >= lvls[3]) & (feeagg_delta.index < lvls[2]),
    (feeagg_delta.index < lvls[3]),
]

df = pd.DataFrame(lvls, columns=['feelvl'])
df.assign(
    feerate=np.ceil(np.exp(df.feelvl/100)*10)/10
    
)

t = pd.DataFrame({
    "MVS": [np.sum(feeagg_delta[ls].mvs) for i, ls in enumerate(lvls_slice)],
    "BTC": [np.sum(feeagg_delta[ls].btc) for i, ls in enumerate(lvls_slice)],
    "AFR": [np.sum(feeagg_delta[ls].btc) / np.sum(feeagg_delta[ls].mvs) * 100 for i, ls in enumerate(lvls_slice)],
    "RFR": np.exp(np.array(lvls/100))
})

t
```
